```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
import numpy as np
```

# Genetic Algorithm

```python
def initialize_population(population_size, features_count):
    return np.random.randint(2, size=(population_size, features_count))

def calculate_fitness(individual, X, y):
    if np.sum(individual) == 0:  # No features selected
        return 0
    selected_features = X[:, individual == 1]
    X_train, X_valid, y_train, y_valid = train_test_split(selected_features, y, test_size=0.2, random_state=42)
    model = LogisticRegression(max_iter=1000)
    model.fit(X_train, y_train)
    predictions = model.predict(X_valid)
    return accuracy_score(y_valid, predictions)

def select_parents(population, fitness, num_parents):
    parents = np.empty((num_parents, population.shape[1]))
    for parent_num in range(num_parents):
        max_fitness_idx = np.where(fitness == np.max(fitness))[0][0]
        parents[parent_num, :] = population[max_fitness_idx, :]
        fitness[max_fitness_idx] = -99999999999
    return parents

def crossover(parents, offspring_size):
    offspring = np.empty(offspring_size)
    crossover_point = np.uint8(offspring_size[1]/2)
    for k in range(offspring_size[0]):
        # Index of the first parent
        parent1_idx = k % parents.shape[0]
        # Index of the second parent
        parent2_idx = (k + 1) % parents.shape[0]
        # The new offspring will have half of its genes from each parent
        offspring[k, 0:crossover_point] = parents[parent1_idx, 0:crossover_point]
        offspring[k, crossover_point:] = parents[parent2_idx, crossover_point:]
    return offspring

def mutation(offspring, mutation_rate):
    # Apply mutation based on mutation rate
    for idx in range(offspring.shape[0]):  # For each offspring
        for gene_idx in range(offspring.shape[1]):  # For each gene in the offspring
            if np.random.rand() < mutation_rate:  # If a randomly chosen number is less than mutation rate
                # Perform mutation (flip the bit)
                offspring[idx, gene_idx] = 1 - offspring[idx, gene_idx]
    return offspring
```

# Parameters

```python
num_generations = 5
population_size = 10 
num_parents_mating = 4 
num_features = x.shape[1]
population = initialize_population(population_size, num_features)
fitness_scores = np.array([calculate_fitness(ind, x.values, y.values) for ind in population])
```

# Evaluation: 

```python
def run_genetic_algorithm(num_generations, population_size, num_parents_mating, mutation_rate):
    # Initialize the population
    population = initialize_population(population_size, num_features)
    
    for generation in range(num_generations):
        print(f"Generation {generation}")

        # Measuring the fitness of each chromosome in the population
        fitness_scores = np.array([calculate_fitness(ind, x.values, y.values) for ind in population])

        # Selecting the best parents for mating
        parents = select_parents(population, fitness_scores, num_parents_mating)

        # Generating the next generation using crossover
        offspring_crossover = crossover(parents, (population_size - parents.shape[0], num_features))

        # Adding some variations using mutation
        offspring_mutation = mutation(offspring_crossover,mutation_rate)

        # Creating the new population based on the parents and offspring
        population[0:parents.shape[0], :] = parents
        population[parents.shape[0]:, :] = offspring_mutation

        # The best result in the current iteration
        print(f"Generation {generation}: Best Fitness {np.max(fitness_scores)}")
        
        

```

# Different Experiments:

```python
experiments = [
    {'num_generations': 5, 'population_size': 10, 'num_parents_mating': 5, 'mutation_rate': 0.01},
    {'num_generations': 5, 'population_size': 20, 'num_parents_mating': 10, 'mutation_rate': 0.05},
    {'num_generations': 5, 'population_size': 50, 'num_parents_mating': 25, 'mutation_rate': 0.1},
]

for exp_num, experiment in enumerate(experiments, start=1):
    print(f"Running Experiment {exp_num}")
    num_generations = experiment['num_generations']
    population_size = experiment['population_size']
    num_parents_mating = experiment['num_parents_mating']
    mutation_rate = experiment['mutation_rate']
    
    run_genetic_algorithm(num_generations, population_size, num_parents_mating, mutation_rate)
    print(f"Experiment {exp_num} complete\n")
```

# Solution

```python
best_fitness_idx = np.argmax(fitness_scores)
best_solution = population[best_fitness_idx, :]
selected_features = x.columns[best_solution == 1]

print("Best solution fitness:", fitness_scores[best_fitness_idx])
print("Selected features:", selected_features)
```
